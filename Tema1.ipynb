### 1. Tipos de Datos en Python

```
# Números enteros
edad = 17
print(type(edad)) # Resultado: <class 'int'>

# Números decimales
pi = 3.14159
print(type(pi)) # Resultado: <class 'float'>

# Cadenas de texto
nombre = "Juan"
print(type(nombre)) # Resultado: <class 'str'>

# Booleanos
es_estudiante = True
print(type(es_estudiante)) # Resultado: <class 'bool'>

# Listas
colores = ["rojo", "verde", "azul"]
print(type(colores)) # Resultado: <class 'list'>

# Tuplas
coordenadas = (10, 20, 30)
print(type(coordenadas)) # Resultado: <class 'tuple'>

# Diccionarios
alumno = {"nombre": "Maria", "edad": 18, "curso": "Matemáticas"}
print(type(alumno)) # Resultado: <class 'dict'>
```

NumPy:

```
import numpy as np

# Crear un array NumPy
datos = np.array([1, 2, 3, 4, 5])
print(datos) # Resultado: [1 2 3 4 5]

# Calcular la media
media = np.mean(datos)
print(media) # Resultado: 3.0

# Calcular la desviación estándar
desviacion_estandar = np.std(datos)
print(desviacion_estandar) # Resultado: 1.5811388300841898
```

### 3. Python para EDA: Explorando los Secretos de los Datos

**Leer y cargar datos:**

```
import pandas as pd

# Leer datos desde un archivo CSV
datos_csv = pd.read_csv("https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv")
```

**Limpiar y preparar datos:**

```
# Detectar valores faltantes
print(datos_csv.isnull().sum()) # Muestra la cantidad de valores faltantes por columna

# Eliminar filas con valores faltantes
datos_csv.dropna(inplace=True)
```

**Describir datos:**

```
# Obtener estadísticas básicas
datos_csv.describe().T # Muestra estadísticas descriptivas de las columnas numéricas
```

**Visualizar datos:**

```
import matplotlib.pyplot as plt

# Crear un histograma
datos_csv["Age"].hist()
plt.show()
```

```
# Crear un diagrama de dispersión
plt.scatter(datos_csv["Age"], datos_csv["Fare"])
plt.xlabel("Edad")
plt.ylabel("Precio del botelo")
plt.show()
```
