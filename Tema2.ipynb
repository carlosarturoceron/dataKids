**Ejemplo: Análisis del conjunto de datos de "iris"**

Exploraremos un conjunto de datos clásico en ciencia de datos, el conjunto de datos de "iris", que contiene información sobre las características físicas de flores de la especie Iris.

**1. Importación de bibliotecas:**

```
import pandas as pd
import matplotlib.pyplot as plt
```

**2. Carga de datos:**

El conjunto de datos Iris es un conjunto de datos clásico en el campo de la ciencia de datos y aprendizaje automático. Se utiliza con frecuencia para enseñar los conceptos básicos de análisis de datos y modelado de machine learning. El conjunto de datos fue creado por el estadístico británico Ronald Fisher en 1936 y se compone de las siguientes características:

Descripción:

Fuente: https://archive.ics.uci.edu/ml/datasets/Iris
Número de muestras: 150
Número de variables: 5
Variables:
SepalLengthCm: Longitud del sépalo en centímetros.
SepalWidthCm: Ancho del sépalo en centímetros.
PetalLengthCm: Longitud del pétalo en centímetros.
PetalWidthCm: Ancho del pétalo en centímetros.
Species: Especie de la flor (Iris setosa, Iris virginica, Iris versicolor).

```
# Carga del conjunto de datos Iris en un DataFrame de pandas
iris = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
```

```
iris.head()
```

**3. Análisis exploratorio de datos:**

**a. Descripción de los datos:**

```
# Descripción estadística del conjunto de datos
iris.describe()
```

**b. Visualización de distribuciones:**

```
# Histograma de la distribución de la característica "SepalLengthCm"
plt.hist(iris["sepal.length"])
plt.xlabel("Longitud del sépalo (cm)")
plt.ylabel("Frecuencia")
plt.show()
```

**c. Gráficos de dispersión:**

```
# Gráfico de dispersión de "SepalLengthCm" vs "SepalWidthCm"
plt.scatter(iris['sepal.length'], iris['sepal.width'])
plt.xlabel("Longitud del sépalo (cm)")
plt.ylabel("Ancho del sépalo (cm)")
plt.show()
```

**4. Modelado de datos:**

**a. Importación de biblioteca de aprendizaje automático:**

```
import statsmodels.api as sm
```

**b. Creación del modelo de regresión lineal:**

```
# Crea un objeto ModelStatsmodels del modelo de regresión lineal
modelo_sm = sm.OLS(iris["sepal.length"], iris[["sepal.width"]]).fit()

# Imprime el resumen del modelo
print(modelo_sm.summary())
```

¡Entender el modelo es sencillo!

**Imagina que "SepalWidthCm" es el tamaño del ancho del sépalo de la flor y "SepalLengthCm" es el tamaño del largo del sépalo.**

El modelo de regresión lineal nos ayuda a **predecir el largo del sépalo (SepalLengthCm)** en base al **ancho del sépalo (SepalWidthCm)**.

**Veamos los resultados clave:**

* **R-squared (0.957):** Cerca de 1, indica que el modelo explica el 95.7% de la variación en el largo del sépalo a partir del ancho.
* **Coeficiente de "SepalWidthCm" (1.8690):** Por cada aumento de 1 unidad en el ancho del sépalo, se espera un aumento promedio de 1.8690 unidades en el largo del sépalo.
* **Valor p de "SepalWidthCm" (0.000):** Muy bajo, significa que la relación entre ancho y largo del sépalo es estadísticamente significativa.

**En resumen, este modelo nos dice que existe una fuerte relación positiva entre el ancho y el largo del sépalo. A mayor ancho, mayor es probable que sea el largo del sépalo.**

**Ten en cuenta que este es un ejemplo simplificado.** En ciencia de datos real, los modelos pueden ser más complejos e involucrar múltiples variables.

¡Espero que esta explicación te sea más clara!

**c. Evaluación del modelo:**

```
from sklearn.metrics import mean_squared_error
import numpy as np

# Realiza predicciones usando el modelo de statsmodels
predicciones = modelo_sm.predict(iris[["sepal.width"]])

# Calcula el error cuadrático medio (MSE)
mse = mean_squared_error(iris["sepal.length"], predicciones)
rmse = np.sqrt(mse)  # Calcula la raíz cuadrada del MSE (RMSE)

# Imprime los resultados
print("Error cuadrático medio (MSE):", mse)
print("Raíz cuadrada del error cuadrático medio (RMSE):", rmse)
```
